```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data extracted from the solutions of the formulations with the input instances
clusters = ["Inferior - Azul", "Inferior - Vermelho", "Superior - Roxo", "Superior - Verde", "Central - laranja"]
vertices = [23, 26, 53, 49, 91]
Q_values = [30, 30, 30, 30, 50]

# Data for F1
UB_F1 = [61807, 89932, 88600, 104339, 153881]
cuts_F1 = [457, 1045, 174, 262, 680]
nodes_F1 = [54576, 32680, 5329, 4707, 997]
time_F1 = [1800, 1800, 1800, 1800, 1800]

# Data for F2
UB_F2 = [56852, 87895, 74808, 92536, 143529]
cuts_F2 = [1775, 346, 672, 305, 1993]
nodes_F2 = [11242, 7151, 6136, 5984, 476]
time_F2 = [62.79, 577.95, 1800, 1800, 1800]
```

```python
np.mean(cuts_F1), np.mean(cuts_F2)
```

```python
def plot_comparisons(x_labels, data_F1, data_F2, y_label, title):
    x = np.arange(len(x_labels))  # Índices para cada cluster
    width = 0.35  # Largura das barras

    fig, ax = plt.subplots()
    bars1 = ax.bar(x - width/2, data_F1, width, label='F1')
    bars2 = ax.bar(x + width/2, data_F2, width, label='F2')

    # Adicionar labels, título e legendas
    ax.set_xlabel('Instância - Cluster')
    ax.set_ylabel(y_label)
    ax.set_title(title)
    ax.set_xticks(x)
    ax.set_xticklabels(x_labels, rotation=45, ha="right")
    ax.legend()

    # Adicionar valores nas barras
    for bars in [bars1, bars2]:
        ax.bar_label(bars, padding=3)

    plt.tight_layout()
    plt.show()
```

```python
plot_comparisons(clusters, cuts_F1, cuts_F2, 'Número de Cortes', 'Comparação de Cortes entre F1 e F2')
```

```python
plot_comparisons(clusters, nodes_F1, nodes_F2, 'Nós Explorados', 'Comparação de Nós Explorados entre F1 e F2')
```

```python
plot_comparisons(clusters, UB_F1, UB_F2, 'Limite Superior (UB)', 'Comparação de UB entre F1 e F2')
```

```python

plot_comparisons(clusters, time_F1, time_F2, 'Tempo (s)', 'Comparação de Tempo entre F1')
```
